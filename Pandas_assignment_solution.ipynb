```python
import os
iris_path = "iris.csv"
if os.path.exists(iris_path):
    iris_df = pd.read_csv(iris_path)
else:
    from sklearn.datasets import load_iris
    sk = load_iris(as_frame=True)
    iris_df = sk.frame
    iris_df['species'] = iris_df['target'].map(dict(enumerate(sk.target_names)))
iris_df.head()
```

```python
iris_copy = iris_df.copy()
if 4 in iris_copy.index:
    iris_mod = iris_copy.drop(index=4)
else:
    iris_mod = iris_copy.drop(iris_copy.index[4])
col_to_drop = iris_mod.columns[3]
iris_mod = iris_mod.drop(columns=[col_to_drop])
iris_mod.head()
```

```python
data_q6 = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Department": ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age": [29, 34, 41, 28, 38],
    "Salary": [50000, 70000, 65000, 55000, 60000],
    "Years_of_Experience": [4, 8, 10, 3, 12],
    "Joining_Date": ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender": ["Female", "Male", "Male", "Female", "Male"],
    "Bonus": [5000, 7000, 6000, 4500, 5000],
    "Rating": [4.5, 4.0, 3.8, 4.7, 3.5]
}
df_q6 = pd.DataFrame(data_q6)
df_q6.shape
df_q6.info()
df_q6.describe()
df_q6.head()
df_q6.tail(3)
df_q6["Salary"].mean()
df_q6["Bonus"].sum()
df_q6["Age"].min()
df_q6["Rating"].max()
df_q6.sort_values(by="Salary", ascending=False)
def cat(r):
    if r >= 4.5:
        return "Excellent"
    elif r >= 4.0:
        return "Good"
    else:
        return "Average"
df_q6["Performance_Category"] = df_q6["Rating"].apply(cat)
df_q6.isnull().sum()
df_q6 = df_q6.rename(columns={"Employee_ID":"ID"})
df_q6[df_q6["Years_of_Experience"] > 5]
df_q6[df_q6["Department"] == "IT"]
df_q6["Tax"] = df_q6["Salary"] * 0.10
df_q6.to_csv("employees_modified.csv", index=False)
df_q6
```
