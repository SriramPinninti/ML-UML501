```python

arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])
print(np.array_equal(arr1, arr2))
```

```python

x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
vals, counts = np.unique(x, return_counts=True)
print(vals[np.argmax(counts)])
print(np.where(x == vals[np.argmax(counts)]))
vals, counts = np.unique(y, return_counts=True)
print(vals[np.argmax(counts)])
print(np.where(y == vals[np.argmax(counts)]))
```

```python

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print(gfg.sum())
print(gfg.sum(axis=1))
print(gfg.sum(axis=0))
```

```python

n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
print(np.trace(n_array))
w, v = np.linalg.eig(n_array)
print(w)
print(v)
print(np.linalg.inv(n_array))
print(np.linalg.det(n_array))
```

```python

p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])
print(np.dot(p1, q1))
print(np.cov(p1, q1))

p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])

print(np.dot(p2, q2))
print(np.cov(p2.T, q2))  # Transpose p2 so rows match for covariance

```

```python

x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
print(np.inner(x, y))
print(np.outer(x, y))
print(np.tensordot(x, y, axes=0))
```

```python

array = np.array([[1, -2, 3],[-4, 5, -6]])
print(np.abs(array))
print(np.percentile(array, [25, 50, 75]))
print(np.percentile(array, [25, 50, 75], axis=0))
print(np.percentile(array, [25, 50, 75], axis=1))
print(np.mean(array))
print(np.median(array))
print(np.std(array))
print(np.mean(array, axis=0))
print(np.median(array, axis=0))
print(np.std(array, axis=0))
print(np.mean(array, axis=1))
print(np.median(array, axis=1))
print(np.std(array, axis=1))
```

```python

a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0])
print(np.floor(a))
print(np.ceil(a))
print(np.trunc(a))
print(np.round(a))
```

```python

array = np.array([10, 52, 62, 16, 16, 54, 453])
print(np.sort(array))
print(np.argsort(array))
print(np.sort(array)[:4])
print(np.sort(array)[-5:])
```

```python

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
ints = array[array == array.astype(int)]
floats = array[array != array.astype(int)]
print(ints)
print(floats)
```

```python

from PIL import Image

def img_to_array(path):
    img = Image.open(path)
    arr = np.array(img)
    if len(arr.shape) == 2:
        np.savetxt("image_grey.txt", arr, fmt='%d')
    else:
        np.savetxt("image_rgb.txt", arr.reshape(-1, arr.shape[-1]), fmt='%d')
    return arr

# Example: arr = img_to_array('your_image_path.jpg')
# Loaded_data = np.loadtxt("image_grey.txt")
```
